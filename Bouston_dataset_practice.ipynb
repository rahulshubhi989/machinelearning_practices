```python
#Practice with the dataset
#load the data

import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/refs/heads/master/BostonHousing.csv')
df
```

```python
# Data Preparation

y = df['tax']
y
```

```python
X = df.drop('tax', axis = 1)
X
```

```python
# Data Splitting

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 100)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
# Model Building
# Linear Regression
# Training the data

from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
y_lr_train_pred = lr.predict(X_train)
y_lr_test_pred = lr.predict(X_test)
```

```python
y_lr_train_pred
```

```python
y_lr_test_pred
```

```python
# Evaluating Model Performance

from sklearn.metrics import mean_squared_error, r2_score

lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2 = r2_score(y_train, y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)
```

```python
print('LR MSE (TRAIN): ', lr_train_mse)
print('LR R2 (TRAIN): ', lr_train_r2)
print('LR MSE (TRAIN): ', lr_test_mse)
print('LR R2 (TRAIN): ', lr_test_r2)
```

```python
lr_results = pd.DataFrame(['Linear Regression', lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]).transpose()
lr_results.columns = ['Method', 'Training MSE', 'TrainingR2', 'Test MSE', 'Test R2']
```

```python
lr_results
```

```python
# Random Forest
# Training the model

from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(max_depth = 2, random_state = 100)
rf.fit(X_train, y_train)
```

```python
# Applying the model

y_rf_train_pred = rf.predict(X_train)
y_rf_test_pred = rf.predict(X_test)
```

```python
# Evaluate Model Performance

from sklearn.metrics import mean_squared_error, r2_score

rf_train_mse = mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2 = r2_score(y_train, y_rf_train_pred)

rf_test_mse = mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2 = r2_score(y_test, y_rf_test_pred)
```

```python
rf_results = pd.DataFrame(['Random Forest', rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]).transpose()
rf_results.columns = ['Method', 'Training MSE', 'TrainingR2', 'Test MSE', 'Test R2']
```

```python
rf_results
```

```python
# Model Comparision

df_models = pd.concat([lr_results, rf_results], axis = 0)
df_models
```

```python
df_models.reset_index(drop = True)
```

```python
# Data Visualization of prediction results

import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize = (5,5))
plt.scatter(x=y_train, y=y_lr_train_pred, c="#7CAE00", alpha = 0.3)

z = np.polyfit(y_train, y_lr_train_pred, 1)
p = np.poly1d(z)
plt.plot(y_train, p(y_train), "#F8766D")
plt.ylabel('predicted tax')
plt.xlabel('Experimental tax')
```
